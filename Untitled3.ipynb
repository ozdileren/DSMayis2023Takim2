```python
def part_of_day(hour):
    if hour>=0 and hour<7:
        return "Night"
    elif hour>=7 and hour<=12:
        return "Morning"
    elif hour>12 and hour<=18:
        return "Afternoon"
    else:
        return "Evening"
```

```python
df2['part_of_day'] = df2.hour.apply(part_of_day)
```

```python
df2["part_of_day"]
```

```python
df2.groupby(['part_of_day', 'day_of_week']).agg({'ridership': 'mean'})
```

```python
partofday_ridership=pd.DataFrame(df2.groupby("part_of_day")["ridership"].mean().sort_values(ascending=False))
```

```python
partofday_ridership
```

```python
plt.bar(partofday_ridership.index,partofday_ridership.ridership,data=partofday_ridership)
plt.xticks(rotation=45,ha="right")
plt.xlabel("Part of day")
plt.ylabel("Ridership")
plt.title("The relationship between part of day and ridership ")
```
